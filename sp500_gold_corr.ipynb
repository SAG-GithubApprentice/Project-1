```python
# Read in the Gold spot price 1 year prediction data
df_Gold_365 = pd.read_csv('forecast_gold_subset.csv')

# Read in the S&P500 close price 1 year prediction data
df_SP500_365 = pd.read_csv('forecast_sp500_subset.csv')
```

```python
correlation = df_Gold_365['yhat'].corr(df_SP500_365['yhat'])
```

```python
# Create a DataFrame to hold the correlation coefficient
correlation_df = pd.DataFrame({'Correlation': [correlation]})

correlation_df.head()
```

```python
import matplotlib.pyplot as plt

# Plot 'trend' column from df_Gold_365
df_Gold_365.plot(x='ds', y='trend', label='Gold', color='gold')

# Plot 'trend' column from df_SP500_365
df_SP500_365.plot(x='ds', y='trend', label='S&P500', color='blue')
```

```python
# Create a scatter plot
plt.scatter(df_Gold_365['yhat'], df_SP500_365['yhat'])

# Set labels and title
plt.xlabel('Gold yhat')
plt.ylabel('S&P500 yhat')
plt.title('Relationship between Gold yhat and S&P500 yhat')

# Show plot
plt.show()
```
